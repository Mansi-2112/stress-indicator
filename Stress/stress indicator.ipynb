```python
import pandas as pd 
import numpy as np
df=pd.read_csv(r'Stress.csv')
print(df.head())
```

```python

x = df[['stress_experience', 'heartbeat_palpitations', 'anxiety_tension', 'sleep_problems', 'restlessness', 'irritability', 'sadness_low_mood', 'loneliness_isolation', 'concentration_problems']]
y = df['stress_type']
print(x, y)
```
